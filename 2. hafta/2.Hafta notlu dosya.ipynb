Kullanıcıdan aldığınız bir sayının "Armstrong" sayısı olup olmadığını bulmaya çalışın.
Örnek olarak, Bir sayı eğer 4 basamaklı ise ve oluşturan rakamlardan herbirinin 4. kuvvetinin toplamı( 3 basamaklı sayılar için 3.kuvveti ) o sayıya eşitse bu sayıya "Armstrong" sayısı denir.

Örnek olarak : 1634 = 1^4 + 6^4 + 3^4 + 4^4

````python
def isanarmstrongnumber(number):
    digits=[int(d)for d in str(number)]
    numberofdigits=len(digits)
    sumofpowers=sum([digit**numberofdigits for digit in digits])

    return sumofpowers==number
number=int(input("enter a number:"))
if isanarmstrongnumber(number):
    print(f"{number} is an armstrong number.")
else:
    print(f"{number} is not an armstrong number.")
````

Kullanıcıdan aldığınız bir sayının mükemmel olup olmadığını bulmaya çalışın.
Bir sayının kendi hariç bölenlerinin toplamı kendine eşitse bu sayıya "mükemmel sayı" denir. Örnek olarak, 6 mükemmel bir sayıdır. (1 + 2 + 3 = 6)

````python
number=int(input("enter a number"))
def isaperfectnumber(number):

    sumofdivisors=0

    for i in range(1, number//2+1):
        if number%i==0:
            sumofdivisors+=i

    return sumofdivisors==number
if isaperfectnumber(number):
    print(f"{number} is a perfect number")
else:
    print(f"{number} is not a perfect number")
````

````python
numbers = [5, 10, 10, 15, 15, 15, 20, 20, 20, 20]
repeatNumbers = []
uniqueNumbers = []

for number in numbers:
    if numbers.count(number) > 1:
        repeatNumbers.append(number)
    elif number not in uniqueNumbers:
        uniqueNumbers.append(number)

# Tekrar etmeyen elemanları bulmak için set kullanıyoruz
uniqueNumbers = [number for number in numbers if numbers.count(number) == 1]

print("Tekrar Eden Elemanlar:", repeatNumbers)
print("Tekrar Etmeyen Elemanlar:", uniqueNumbers)
````

````python
from math import*
def factorial(sayi):
    print("kendi faktöriyel fonksiyonum")
    faktoriyel=1
    if(sayi==0 or sayi==1):
        return 1
    while(sayi>=1):
        faktoriyel *=sayi
        sayi-=1
    return faktoriyel
factorial(5)
````

Liderlerimizin kodlamak istediği oyun kuralları şöyledir:
- Oyunda random olarak 0 ile 100 arasında bir sayı tutulur.
- Kullanıcıdan bilgisayarın tuttuğu sayıyı tahmin etmesi için bir sayı girmesini istenir.
- Eğer kullanıcının girdiği sayı bulunması gereken sayıdan küçükse  tahmininizi yükseltin, büyükse tahmininizi düşürün şeklinde bir çıktı verilir.
- Oyunun gerçeğe yakın olması için bilgisayarın çalışmasını simgelemek için 1 saniye bekleme süresi tanınır.
- Kullanıcı 5 hakkı vardır. 5 hakkı da kullanıp sayıyı bulamazsa oyunu kaybeder.

````python
import time
import random

dogrusayi=random.randint(1,100)

haklar=5

while haklar>0:
    kullanicisayi=int(input("1-100 arası bir sayı gir:"))

    if kullanicisayi<dogrusayi:
        haklar-=1
        if haklar>0:
            print(f"tahminini yükselt.{haklar} hakkın kaldı.")

        else:
            print("oyunu kaybettin.")

    elif kullanicisayi>dogrusayi:
        haklar-=1
        if haklar>0:
            print(f"tahminini düşür.{haklar} hakkın kaldı.")

        else:
            print("oyunu kaybettin.")

    else:
        print("tebrikler!doğru tahmin ettiniz.")
        break
    
    if haklar>0:
        print("bekle...")
        time.sleep(1)
        print("1 saniye sonra")

if haklar==0:
    print(f"doğru sayı {dogrusayi} idi.")
````

2. Diyet Planı Oluşturma

Proje Adı: Diyet Planı Oluşturma

Açıklama:

	•	Kullanıcının yaş, cinsiyet, boy, kilo ve aktivite seviyesine göre kişiselleştirilmiş diyet planı oluşturun.
	•	Kullanıcının günlük kalori ihtiyacını hesaplayın ve önerilerde bulunun.

Adımlar:

	1.	Veri Toplama: Kullanıcı verilerini alın.
	2.	Veri Temizleme: Eksik veya hatalı verileri düzeltin.
	3.	Kalori Hesaplama: Kullanıcının günlük kalori ihtiyacını hesaplayın.
	4.	Diyet Planı Oluşturma: Kullanıcının ihtiyacına göre yiyecekler önerin.

Örnek Kod:

````python
def calculate_calories(age, gender, weight, height, activity_level):
    if gender == 'male':
        bmr = 10 * weight + 6.25 * height - 5 * age + 5
    else:
        bmr = 10 * weight + 6.25 * height - 5 * age - 161
    activity_multiplier = {'sedentary': 1.2, 'light': 1.375, 'moderate': 1.55, 'active': 1.725}
    return bmr * activity_multiplier[activity_level]

def create_diet_plan(calories):
    foods = {
        'breakfast': [('oatmeal', 150), ('egg', 78), ('banana', 105)],
        'lunch': [('chicken breast', 165), ('quinoa', 120), ('broccoli', 55)],
        'dinner': [('salmon', 200), ('sweet potato', 112), ('asparagus', 40)],
        'snack': [('almonds', 160), ('apple', 95), ('yogurt', 59)]
    }
    plan = {}
    for meal, items in foods.items():
        meal_calories = sum([item[1] for item in items])
        portion = calories / (meal_calories * 4)
        plan[meal] = [(item[0], item[1] * portion) for item in items]
    return plan

# Kullanıcı bilgileri
age = 25
gender = 'female'
weight = 70  # kg
height = 170  # cm
activity_level = 'moderate'

# Kalori ihtiyacını hesapla
calories_needed = calculate_calories(age, gender, weight, height, activity_level)

# Diyet planı oluştur
diet_plan = create_diet_plan(calories_needed)
print(diet_plan)
````

Tabii ki! Bu diyet planı projesini daha kapsamlı ve ilgi çekici hale getirmek için birkaç ek özellik ekleyebiliriz:

	1.	Farklı Diyet Tipleri Ekleyin:
	•	Vegan, vejetaryen, düşük karbonhidrat gibi farklı diyet türlerini destekleyin.
	2.	Besin Değerlerini Hesaplayın:
	•	Kaloriye ek olarak, protein, karbonhidrat, yağ gibi besin değerlerini de hesaplayın.
	3.	Kullanıcı Geri Bildirimi ve Ayarlamalar:
	•	Kullanıcının diyet planını değerlendirmesine olanak tanıyın ve gerektiğinde planı ayarlayın.
	4.	Haftalık Plan Oluşturun:
	•	Tek günlük plan yerine haftalık planlar oluşturun.
	5.	Rastgele Yemek Önerileri:
	•	Her yemek için rastgele yemek önerileri sunun.

İşte bu ek özelliklerle daha detaylı bir proje örneği:

````python
import numpy as np

def calculate_calories(age, gender, weight, height, activity_level):
    if gender == 'male':
        bmr = 10 * weight + 6.25 * height - 5 * age + 5
    else:
        bmr = 10 * weight + 6.25 * height - 5 * age - 161
    activity_multiplier = {'sedentary': 1.2, 'light': 1.375, 'moderate': 1.55, 'active': 1.725}
    return bmr * activity_multiplier[activity_level]

def create_diet_plan(calories, diet_type='standard'):
    foods = {
        'standard': {
            'breakfast': [('oatmeal', 150, 5, 27, 3), ('egg', 78, 6, 1, 5), ('banana', 105, 1, 27, 0)],
            'lunch': [('chicken breast', 165, 31, 0, 3), ('quinoa', 120, 4, 21, 2), ('broccoli', 55, 4, 11, 0)],
            'dinner': [('salmon', 200, 22, 0, 13), ('sweet potato', 112, 2, 26, 0), ('asparagus', 40, 2, 7, 0)],
            'snack': [('almonds', 160, 6, 6, 14), ('apple', 95, 0, 25, 0), ('yogurt', 59, 10, 3, 0)]
        },
        'vegan': {
            'breakfast': [('smoothie', 200, 5, 50, 1), ('chia pudding', 180, 6, 20, 9)],
            'lunch': [('tofu stir-fry', 250, 20, 30, 10), ('quinoa salad', 200, 6, 35, 5)],
            'dinner': [('lentil soup', 250, 18, 40, 8), ('vegan burger', 300, 22, 45, 12)],
            'snack': [('hummus & carrots', 150, 4, 20, 8), ('fruit salad', 100, 1, 25, 0)]
        }
    }
    
    chosen_foods = foods[diet_type]
    plan = {}
    for meal, items in chosen_foods.items():
        meal_calories = sum([item[1] for item in items])
        portion = calories / (meal_calories * 4)
        plan[meal] = [(item[0], item[1] * portion, item[2] * portion, item[3] * portion, item[4] * portion) for item in items]
    return plan

def summarize_diet_plan(plan):
    summary = {}
    for meal, items in plan.items():
        total_calories = sum([item[1] for item in items])
        total_protein = sum([item[2] for item in items])
        total_carbs = sum([item[3] for item in items])
        total_fat = sum([item[4] for item in items])
        summary[meal] = {
            'calories': total_calories,
            'protein': total_protein,
            'carbs': total_carbs,
            'fat': total_fat
        }
    return summary

# Kullanıcı bilgileri
age = 25
gender = 'female'
weight = 70  # kg
height = 170  # cm
activity_level = 'moderate'
diet_type = 'vegan'  # standard veya vegan seçilebilir

# Kalori ihtiyacını hesapla
calories_needed = calculate_calories(age, gender, weight, height, activity_level)

# Diyet planı oluştur
diet_plan = create_diet_plan(calories_needed, diet_type)

# Diyet planı özetini oluştur
diet_summary = summarize_diet_plan(diet_plan)

print("Diyet Planı:")
for meal, items in diet_plan.items():
    print(f"{meal.capitalize()}:")
    for item in items:
        print(f"  - {item[0]}: {item[1]:.2f} kcal, Protein: {item[2]:.2f}g, Karbonhidrat: {item[3]:.2f}g, Yağ: {item[4]:.2f}g")

print("\nDiyet Özeti:")
for meal, nutrients in diet_summary.items():
    print(f"{meal.capitalize()}: {nutrients['calories']:.2f} kcal, Protein: {nutrients['protein']:.2f}g, Karbonhidrat: {nutrients['carbs']:.2f}g, Yağ: {nutrients['fat']:.2f}g")
````

Yeni Eklenen Özellikler:

	1.	Farklı Diyet Tipleri: diet_type parametresiyle vegan ve standart diyetler arasında seçim yapılabilir.
	2.	Besin Değerleri: Her öğün için kalori, protein, karbonhidrat ve yağ miktarları hesaplanır.
	3.	Haftalık Plan: Tek günlük plan yerine haftalık plan oluşturma fikri eklenebilir.
	4.	Rastgele Yemek Önerileri: Yemekler rastgele seçilebilir.

5. Basit Şifreleme ve Şifre Çözme

Proje Adı: Basit Şifreleme ve Şifre Çözme

Açıklama:

	•	Kullanıcının girdiği metni basit bir şifreleme algoritması ile şifreleyin ve çözün.

Adımlar:

	1.	Şifreleme Fonksiyonu: Basit bir şifreleme algoritması yazın.
	2.	Şifre Çözme Fonksiyonu: Şifrelenmiş metni çözen bir fonksiyon yazın.
	3.	Kullanıcı Girdisi: Kullanıcıdan metin ve şifreleme anahtarı alın.
	4.	Sonuç Gösterme: Şifrelenmiş ve çözülmüş metni ekranda gösterin.

Örnek Kod:

````python
def sifreleme(metin, anahtar):
    sifreli_metin = ""
    for karakter in metin:
        sifreli_metin += chr(ord(karakter) + anahtar)
    return sifreli_metin

def sifre_cozme(sifreli_metin, anahtar):
    cozulmus_metin = ""
    for karakter in sifreli_metin:
        cozulmus_metin += chr(ord(karakter) - anahtar)
    return cozulmus_metin

metin = input("Şifrelenecek metni girin: ")
anahtar = int(input("Şifreleme anahtarını girin (sayı): "))

sifreli_metin = sifreleme(metin, anahtar)
print(f"Şifreli Metin: {sifreli_metin}")

cozulmus_metin = sifre_cozme(sifreli_metin, anahtar)
print(f"Çözülmüş Metin: {cozulmus_metin}")
````

````python
import csv
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Veriyi yükle
def load_data(file_name):
    with open(file_name, newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        data = list(reader)
    return data

ratings = load_data('ratings.csv')

# Kullanıcı-film matrisi oluştur
user_ids = list(set([int(row[0]) for row in ratings[1:]]))
movie_ids = list(set([int(row[1]) for row in ratings[1:]]))

user_movie_matrix = np.zeros((len(user_ids), len(movie_ids)))

for row in ratings[1:]:
    user_id = int(row[0])
    movie_id = int(row[1])
    rating = float(row[2])
    user_movie_matrix[user_ids.index(user_id)][movie_ids.index(movie_id)] = rating

# Benzerlik matrisini hesapla
user_similarity = cosine_similarity(user_movie_matrix)
user_similarity[user_similarity < 0] = 0  # Negatif benzerlikleri sıfıra ayarla

# Kullanıcı için film önerisi yap
def recommend_movies(user_id, num_recommendations):
    user_index = user_ids.index(user_id)
    similar_users = np.argsort(user_similarity[user_index])[::-1]
    recommended_movies = {}
    
    for similar_user in similar_users:
        if similar_user == user_index:
            continue
        for movie_index, rating in enumerate(user_movie_matrix[similar_user]):
            if user_movie_matrix[user_index][movie_index] == 0 and rating > 0:
                if movie_index not in recommended_movies:
                    recommended_movies[movie_index] = rating
                else:
                    recommended_movies[movie_index] += rating
        if len(recommended_movies) >= num_recommendations:
            break
    
    recommended_movie_indices = sorted(recommended_movies, key=recommended_movies.get, reverse=True)
    return [movie_ids[index] for index in recommended_movie_indices[:num_recommendations]]

# Önerileri yazdır
print(recommend_movies(user_id=1, num_recommendations=5))
````

	3.	Özellik Mühendisliği: Kullanıcı-film matrisi oluşturun.
	4.	Model Eğitim: Basit bir kullanıcı tabanlı işbirlikçi filtreleme algoritması kullanın.
	5.	Öneri Yapma: Kullanıcıların beğenebileceği filmleri önerin.

````python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Veriyi yükle
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

# Kullanıcı-film matrisi oluştur
user_movie_matrix = ratings.pivot(index='userId', columns='movieId', values='rating')

# Benzerlik matrisini hesapla
user_similarity = cosine_similarity(user_movie_matrix.fillna(0))
user_similarity_df = pd.DataFrame(user_similarity, index=user_movie_matrix.index, columns=user_movie_matrix.index)

# Kullanıcı için film önerisi yap
def recommend_movies(user_id, num_recommendations):
    similar_users = user_similarity_df[user_id].sort_values(ascending=False).index[1:]
    recommended_movies = pd.Series()
    for similar_user in similar_users:
        movies_watched = user_movie_matrix.loc[similar_user].dropna().index
        recommended_movies = recommended_movies.append(movies_watched)
        if len(recommended_movies) >= num_recommendations:
            break
    return recommended_movies.value_counts().index[:num_recommendations]

# Önerileri yazdır
print(recommend_movies(user_id=1, num_recommendations=5))
````
