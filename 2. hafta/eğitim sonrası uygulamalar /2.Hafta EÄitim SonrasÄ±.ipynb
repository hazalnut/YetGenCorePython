### Soru 7
Sen ve YetGen ekibin bir restorana gidiyorsunuz. Garson size iyi davranıyor, bu yüzden ona bahşiş vermeye karar veriyorsunuz. Gelen hesap üzerinden %12 oranında bahşiş bırakmak ve hesabı eşit olarak bölüşmek istiyorsunuz. Kişi başı ne kadar hesap ödemeniz gerekir?

İpucu: Kişi sayısı, fatura ve bahşiş oranını input ile alabilirsiniz.

### Soru 8
YetGen Harikalar Diyarın'da;

Her 7 saniyede bir doğum gerçekleşiyor.

Her 13 saniyede bir ölüm meydana geliyor.

Her 45 saniyede bir beyin göçü alınıyor.

Önümüzdeki 5 yıl sonraki popülasyonu hesaplamak için bir program yazın. Şu anki YetGen Harikalar Diyarı nüfusunun 2.000.000 kişi, 1 yılda 365 gün olduğunu varsayın.

İpucu : Pythonda bölme işlemi için tam sayı bölme operatörünü (//) kullanabilirsiniz. Örneğin; 5//4=1 (1.25 değil), 10//4=2 (2.5 değil) 

**Örnek**
```
>>>input
2000000
>>>output
15900484
```


````python
saniye=int(365*24*60*60*5)
dogum=int(saniye//7)
olum=int(saniye//13)
beyingocu=int(saniye//45)
nufus=2000000
populasyon=int(nufus+dogum+beyingocu-olum)
print(populasyon)
````
