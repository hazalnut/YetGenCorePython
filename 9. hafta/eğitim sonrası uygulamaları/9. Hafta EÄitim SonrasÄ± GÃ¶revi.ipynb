Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import time 
import random

class Kumanda:
    def __init__(self,name:str,volume:int,tv_status:bool,kanal_listesi:list,kanal:str):
        self.name=name
        self.volume=volume
        self.tv_status=tv_status
        self.kanal_listesi=kanal_listesi
        self.kanal=kanal
    
    def tv_ac(self):
        if (self.tv_status==False):
            self.tv_status=True
            print("TV açılıyor")
        else:
            print("TV zaten açık")
    
    def tv_kapat(self):
        print("TV kapanıyor")
        self.tv_status==False

    def ses_ayarla(self):
        while True:
            inp=input("sesi azalt:<\n sesi artır:>\n çıkış:q")
            if inp=="q":
                break
            elif inp=="<" and self.volume>0:
                self.volume-=1
            elif inp==">" and self.volume<100:
                self.volume+=1
            elif inp=="<<" and self.volume>4:
                self.volume-=5
            elif inp==">>" and self.volume<96:
                self.volume+=5
            print("ses:", self.volume)
        print("ses ayarı bitti. güncel ses durumu:"+ str(self.volume))

    def kanal_ekle(self,kanal):
        self.kanal_listesi.append(kanal)
        print(f"{kanal} kanalı listeye eklendi.")

    def rastgele_kanal(self):
        rastgele=random.randindt(0,len(self.kanal_listesi)-1)
        self.kanal=self.kanal_listesi[rastgele]
        print("şu anki kanal:", self.kanal)

    def __len__(self):
        return len(self.kanal_listesi)

    def __str__(self):
        return (f"tv durumu:{self.tv_status}\nkanal listesi:{self.kanal_listesi}\ngüncel kanal:{self.kanal}\ntv sesi:{self.volume}")
    
kumanda=Kumanda()

print("""
Televizyon Uygulaması
1.TV aç
2.TV kapat  
3.ses ayarları     
4.kanal ekle      
5.kanal sayısını öğrenme      
6.rastgele kanala geçme
7.televizyon bilgileri
çıkmak için 'q' ya basın.
""")

while True:
    inp=input("işlem girin:")
    if inp=='q':
        print("kumandayı camdan attın ya da kapattın")
        break
    elif inp=='1':
        kumanda.tv_ac()
    elif inp=='2':
        kumanda.tv_kapat()
    elif inp=='3':
        kumanda.ses_ayarla()
    elif inp=='4':
        kanal_isim=input("kanal ismi girin:")
        kumanda.kanal_ekle(kanal_isim)
    elif inp=='5':
        print("kanal sayısı:",len(kumanda))
    elif inp=='6':
        kumanda.rastgele_kanal()
    elif inp=='7':
        print(kumanda)
```
