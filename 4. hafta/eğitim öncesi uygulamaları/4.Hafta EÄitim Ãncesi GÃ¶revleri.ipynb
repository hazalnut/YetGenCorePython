# Problem 1
Sözlükleri kullanarak bir telefon rehberi yazın. Bu rehberde kullanıcıya kimin telefonunu görüntülemek istediğini sorun ve kullanıcının girdiği isme göre o kişinin telefon numarasını yazdırın. Proje sonunda elde edeceğiniz çıktı  şuna benzer olmalı:


Lütfen numarasını öğrenmek istediğiniz kişinin adını girin: Ahmet

````python
phonebook={"hazal":"05392466003","evrim":"05555752244","sude":"05464961926","mirza":"05334009059"}
name=input("please enter the name of the person whose number you want to know.")
if name in phonebook:
    print(f"phone number of {name} is:{phonebook[name]}")
else:
    print(f"this phone number can not be found in phonebook.")
````

Lütfen bilgilerini görüntülemek istediğiniz çalışanın ismini girin: Mehmet Yağız

````python
officeemployees={"hazal kocaoğlu":["hatay",20,"engineer"],"sude naz çam":["bodrum",21,"engineer"],"evrim cantimur":["malatya",42,"advertiser"]}
name=input("enter the name of the employee you want to view.")
if name in officeemployees:
    hometown,age,department=officeemployees[name]
    print(f"{name}=hometown:{hometown} age:{age} department:{department}")
else:
    print("there is no employee in the name you searched for.")
````
